# 1.What is Hypothesis Testing?
`statistical method used to make conclusion(inferences)`
- Hypothesis testing can be defined as a statistical tool that is used to identify if the results of an experiment are meaningful or not. It involves setting up a null hypothesis and an alternative hypothesis. These two hypotheses will always be mutually exclusive. This means that if the null hypothesis is true then the alternative hypothesis is false and vice versa.<br>
- Hypothesis testing uses sample data from the population to draw useful conclusions regarding the population probability distribution. It tests an assumption made about the data using different types of hypothesis testing methodologies.The hypothesis testing results in either rejecting or not rejecting the null hypothesis.<br>

- Hypothesis testing provides a way to verify whether the results of an experiment are valid. <br>

Making data-driven decisions and evaluating the validity of scientific claims based on sample evidence.

## Hypothesis Testing P Value
In hypothesis testing, the p value is used to indicate whether the results obtained after conducting a **test are statistically significant or not.** It also indicates the probability of making an error in rejecting or not rejecting the null hypothesis.This value is always a number between 0 and 1. The p value is compared to an alpha level,  or significance level. The alpha level can be defined as the acceptable risk of incorrectly rejecting the null hypothesis. The alpha level is usually chosen between 1% to 5%.

## Two-sample T-test With Python
- The Independent Samples t Test or 2-sample t-test compares the means of two independent groups in order to determine whether there is statistical evidence that the associated population means are significantly different. The Independent Samples t Test is a parametric test. This test is also known as: **Independent t Test**
**_,p_value=stats.`ttest_ind`(`a=classA_height,b=ClassB_ages`,equal_var=False)**

### Example: Two-Sample t-Test in Python
we have two sets of data: blood pressure measurements from a 
- Group taking the new drug and a 
- Group taking a placebo.

#### 1.Formulate Hypotheses:

0: The new drug has no effect on blood pressure (mean difference = 0).

1: The new drug lowers blood pressure (mean difference < 0).

#### 2.Select Significance Level:
-  = 0.05
#### 3.Choose the Test:
- We'll use a two-sample t-test.
#### 4.Calculate the Test Statistic:
- This will be done using Python libraries.
#### 5.Determine the p-value:
- We'll calculate this from the t-distribution.
#### 6.Make a Decision: 
- Compare the p-value to .

```python
import numpy as np
from scipy import stats

# Sample data
np.random.seed(42)  # for reproducibility
drug_group = np.random.normal(loc=120, scale=10, size=30)  # mean=120, std=10
placebo_group = np.random.normal(loc=130, scale=10, size=30)  # mean=130, std=10

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(drug_group, placebo_group, alternative='less')

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision based on significance level
alpha = 0.05
# 1.588 > 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new drug lowers blood pressure.")
else:
    print("Fail to reject the null hypothesis: No sufficient evidence to say the drug lowers blood pressure.")
```

## One-sample T-test with Python
- The test will tell us whether means of the sample and the population are different
OR<br>
- A one-sample t-test compares the mean of a single sample to a known value (population mean).
**_,p_value=stats.`ttest_1samp`(`a=classA_ages,popmean=school_ages.mean()`)**



Let's go through an example of a one-sample t-test in Python. In this example, we will test if the mean blood pressure of a group of patients after taking a new drug is significantly different from a known population mean (let's say the average blood pressure of the general population is 130 mmHg).

- *Hypotheses Null Hypothesis ( 0)*: The mean blood pressure of patients taking the new drug is equal to 130 mmHg (=130).
<br>
- *Alternative Hypothesis ( 1)*: The mean blood pressure of patients taking the new drug is not equal to 130 mmHg (130).

**1. Example One-sample T-test with Python**

```python
import numpy as np
from scipy import stats

# Sample data
np.random.seed(42)  # for reproducibility
blood_pressure_data = np.random.normal(loc=125, scale=10, size=30)  # mean=125, std=10

# Population mean
population_mean = 130

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(blood_pressure_data, population_mean)

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision based on significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The mean blood pressure is significantly different from 130 mmHg.")
else:
    print("Fail to reject the null hypothesis: No sufficient evidence to say the mean blood pressure is different from 130 mmHg.")
```

## Paired T-test With Python
When you want to check how different samples from the same group are, you can go for a paired T-test OR <br>
To compare the means of two related samples. It checks if there is a significant difference between paired observations.


**1. Example Paired T-test With Python**

You have weights of individuals before and after a diet program and want to see if the program has a significant effect.

```python
from scipy import stats

# Sample data (weights before and after the diet program)
weight_before = np.array([65, 70, 75, 80, 85])
weight_after = np.array([64, 69, 73, 79, 84])
```

```python
# Perform the paired t-test
t_statistic, p_value = stats.ttest_rel(weight_before, weight_after)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

## Chi-Square Test
- The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.

```python
import scipy.stats as stats
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
dataset=sns.load_dataset('tips')
```

```python
dataset.head()
```

```python
dataset_table=pd.crosstab(dataset['sex'],dataset['smoker'])
print(dataset_table)
```

```python
dataset_table.values 
```

```python
#Observed Values
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
```

```python
val=stats.chi2_contingency(dataset_table)
```

```python
val
```

```python
Expected_Values=val[3]
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
